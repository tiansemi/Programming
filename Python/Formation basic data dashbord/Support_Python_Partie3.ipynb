## 1. Notre outil principal : Pandas
Nous utilisons la bibliothèque **Pandas** pour manipuler nos tableaux (appelés "DataFrames").

```python
import pandas as pd
print("Pandas est prêt pour nos analyses !")
```

## 2. Importer nos fichiers
Nous gérons ici les deux formats les plus courants : le **CSV** (avec séparateur `,`) et le **Excel**.

```python
# Nous précisons le séparateur virgule pour ce fichier spécifique
df = pd.read_csv("ventes.csv", sep=",")

# Import Excel
df_excel = pd.read_excel("ventes.xlsx", sheet_name="Feuil1")

print("Nos structures d'importation sont prêtes.")
```

## 3. Explorer nos données
Une fois le fichier chargé, nous vérifions son contenu : taille, colonnes (city, total_price...) et types de données.

```python
print(df.head())      # Nous voyons les 5 premières lignes
print(df.shape)     # Nous voyons le nombre de (lignes, colonnes)
print(df.info())    # Nous vérifions les colonnes et valeurs vides
print(df.dtypes)    # Nous vérifions si 'total_price' est bien numérique
```

## 4. Notre contrôle qualité
Nous traquons les erreurs classiques : les cases vides et les lignes en double.

```python
# Compter nos valeurs manquantes par colonne
print(df.isnull().sum())

# Identifier nos doublons
print(f"Nombre de doublons détectés : {df.duplicated().sum()}")
```

## 5. Filtrage et sélection
Nous isolons uniquement les indicateurs qui nous intéressent.

```python
# Sélectionner des colonnes spécifiques
notre_selection = df[["product_name", "city", "total_price"]]

# Filtrer nos lignes (ex: ventes > 200)
gros_montants = df[df["total_price"] > 200]

# Filtrer par condition multiple (ex: Membres à Chicago)
membres_chicago = df[(df["city"] == "Chicago") & (df["customer_type"] == "Member")]
```

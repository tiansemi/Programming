## 1. Premier pas : l'affichage

## 2. Nos variables et types de données
Nous stockons des informations dans des "boîtes" (variables).
* **Texte** : entre guillemets `" "` (String)
* **Entier** : nombre sans virgule (Integer)
* **Décimal** : nombre avec un **point** (Float)
* **Booléen** : vrai ou faux (`True`/`False`)

```python
logiciel = "SAGE"
quantite = 10
prix_unitaire = 99.90
est_valide = True

print("Nous utilisons", logiciel, "pour un montant de :", quantite * prix_unitaire)
```

## 3. Nos calculs de gestion
Nous utilisons les opérateurs classiques : `+`, `-`, `*`, `/`.

```python
ca_janvier = 100000
ca_fevrier = 125000

croissance = ((ca_fevrier - ca_janvier) / ca_janvier) * 100
print(f"Notre croissance : {croissance}%")
```

## 4. Nos exercices pratiques
Appliquons nos bases sur ces 3 cas :

```python
# Exercice 1 : Calculez notre total de ventes
v1, v2 = 120000, 80000
total = v1 + v2
print("Total :", total)

# Exercice 2 : Calculez notre résultat (CA - Dépenses)
ca, depenses = 250000, 120000
print("Résultat :", ca - depenses)

# Exercice 3 : Simulez une ligne de facture
client = "Entreprise A"
montant_ht = 1500
print(f"Client {client} | TTC : {montant_ht * 1.20}")
```

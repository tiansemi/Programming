## 2. Nos structures de données
Nous ne manipulons pas qu'une seule valeur, mais des ensembles.

* **La Liste `[]`** : pour stocker une suite de montants (ex: ventes du mois).
* **Le Dictionnaire `{}`** : pour structurer des informations (ex: fiche client).

```python
# Nos listes
ventes = [15000, 20000, 18000]
ventes.append(25000) # Nous ajoutons une vente
print(f"Première vente : {ventes[0]} | Nombre de ventes : {len(ventes)}")

# Nos dictionnaires
notre_client = {
    "nom": "Entreprise A",
    "ville": "Abidjan",
    "ca": 500000
}
print(f"Client : {notre_client['nom']} ({notre_client['ville']})")
```

## 3. Automatisation : Boucles et Conditions
* **La boucle `for`** : pour parcourir nos données sans effort.
* **La condition `if/else`** : pour prendre des décisions automatiques.

```python
ventes = [15000, 22000, 18000, 25000]
total = 0
seuil = 20000

for v in ventes:
    total += v
    if v > seuil:
        print(f"Vente {v} : Objectif dépassé")
    else:
        print(f"Vente {v} : Sous le seuil")

print(f"--- Total cumulé : {total} ---")
```

## 4. Nos fonctions
Nous créons nos propres outils pour éviter de répéter les mêmes calculs.

```python
def analyser_marge(revenu, cout):
    marge = revenu - cout
    taux = (marge / revenu) * 100
    return marge, taux

# Utilisation de notre fonction
m, t = analyser_marge(200000, 150000)
print(f"Marge : {m} | Taux de marge : {t}%")
```

## 5. Synthèse finale : Notre script d'analyse
Regroupons tout ce que nous avons appris.

```python
ventes = [12000, 15000, 18000]
total = sum(ventes)
moyenne = total / len(ventes)

message = "Performance correcte" if moyenne > 14000 else "À surveiller"

print(f"Synthèse -- Moyenne : {moyenne} | Diagnostic : {message}")
```
